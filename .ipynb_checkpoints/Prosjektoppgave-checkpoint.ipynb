```python
import numpy as np
from matplotlib import pyplot as plt

# Av Jonathan Talseth og Magnus Krane


# Bolig verdi = 43000000
# Egenkapital 15% = 645000
# Renta 1.9 % = 0.019 %
# Nedbetalingstid 25 år 


#Regner ut egenkapital fra prosent
Bolig_verdi = 4300000
Egenkapital_pros = 0.15
Renta = 0.019
Terminer = 25

# X verdi kan endres for å se utregning av ulike terminer 
x = (25)

#Egenkapitalen for å ta opp et boliglån på 4300000 er minimum 15%. Regner ut prosent
Bolig_verdi = 4300000
Egenkapital_pros = 0.15

Egenkapital = Bolig_verdi*Egenkapital_pros
print("Egenkapital: " + str(Egenkapital))

#Regner ut hvor mye vi må låne fra banken etter egenkapitalen er betalt
Bolig_lån = Bolig_verdi - Egenkapital
print("Huslån: " + str(Bolig_lån))


#Terminbeløp

Terminbeløp = (3655000*Renta*(1+Renta)**Terminer)/((1+Renta)**Terminer-1)
print("Terminbeløp: " + str(Terminbeløp))

#Sum av lån

Sum_lån =Terminbeløp*Terminer

print("Total lånesum: " + str(Sum_lån))

#Rente per termin
Rente = (3655000*Renta)/x

print("Rente for " + str(x) + " termin : kr " + str(Rente))


#Avdrag for terminer

Avdrag = Terminbeløp-Rente

print("Avdrag for " + str(x) + " termin : kr " + str(Avdrag))


# Verdi for rente
R1 = (3655000*0.019)/1
R5 = (3655000*0.019)/5
R10 = (3655000*0.019)/10
R15 = (3655000*0.019)/15
R20 = (3655000*0.019)/20
R25 = (3655000*0.019)/25


#Presenterer ulike terminer i et stolpediagram

plt.style.use('ggplot')
x = ['Termin 1', 'Termin 5', 'Termin 10', 'Termin 15', 'Termin 20','Termin 25']
g = [R1, R5, R10 ,R15, R20, R25]
x_pos = np.arange(len(x))
plt.bar(x_pos, g, color=['#7ed6df', 'magenta', 'lawngreen', 'red', 'yellow', 'darkorange'])
plt.xlabel("Termin")
plt.ylabel("Rente")
plt.title("Annuitetslån")
plt.xticks(x_pos, x)

plt.show()


#Presenterer ulike terminer i et kakediagram 

labels = 'Termin1', 'Termin5', 'Termin10', 'Termin15', 'Termin20', 'Termin25'
sizes = [R1, R5, R10, R15, R20, R25]
explode = (0, 0, 0, 0, 0.3, 0.5) #

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')


plt.show()

```
